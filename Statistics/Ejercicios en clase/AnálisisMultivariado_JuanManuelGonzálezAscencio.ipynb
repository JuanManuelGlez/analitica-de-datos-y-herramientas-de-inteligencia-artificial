```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset('mpg')
dataset.head()
```

```python
# Creamos Pairplot de las variables numéricas del dataset

sns.pairplot(dataset)
```

```python
plt.figure(dpi=120) # resolution
sns.pairplot(dataset, hue='mpg') # genera una matriz con plots de dispersion
plt.show()
```

## Matriz de calor sintetizada

```python
# Correlaciones
plt.figure(dpi=120, figsize=(5,5)) # resolution
mask = np.triu(np.ones_like(dataset.corr(numeric_only=True),dtype = bool))

sns.heatmap(dataset.corr(numeric_only=True),lw=1,cmap = 'plasma', mask=mask,  annot=True)
plt.title('Correlation Heatmap')
```

```python
#Quitamos las columnas que no sean numeéricas

dataset = dataset.select_dtypes(include=['float64', 'int64'])
dataset.head()
```

```python
#Jointplot
print("Jointplot de mpg y las demas variables")

for i in dataset.columns:
    if i != 'mpg':
        print(f"Jointplot de mpg y {i}")
        sns.jointplot(x = i, y = 'mpg', data = dataset, kind = 'reg')
        plt.show()
```

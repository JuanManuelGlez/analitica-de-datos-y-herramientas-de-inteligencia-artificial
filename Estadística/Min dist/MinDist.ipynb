## Separamos 135 patrones para la clase entrenamiento y dejamos 15 para la clase test

```python

X_train, X_test = train_test_split(iris, test_size = 0.1, random_state = 42)
```

```python
X_train
```

sacamos los vectores con los promedios

```python
#Creamos tres listas para cada tipo de flor con los promedios de cada columna

Iris_versicolor = []
Iris_virginica = []
Iris_setosa= []

for i in range(4):
    Iris_versicolor.append(X_train[X_train['class'] == 'Iris-versicolor'].iloc[:,i].mean())
    Iris_virginica.append(X_train[X_train['class'] == 'Iris-virginica'].iloc[:,i].mean())
    Iris_setosa.append(X_train[X_train['class'] == 'Iris-setosa'].iloc[:,i].mean())

Iris_versicolor
```

Calculamos la distancia euclidiana entre cada flor y los promedios de cada clase

```python
def euclidean_distance(x1, x2):
    distance = 0
    for i in range(len(x1)):
        distance += pow((x1[i] - x2[i]), 2)
    return math.sqrt(distance)

def distance_to_versicolor(x):
    return euclidean_distance(x, Iris_versicolor)   

def distance_to_virginica(x):
    return euclidean_distance(x, Iris_virginica)

def distance_to_setosa(x):
    return euclidean_distance(x, Iris_setosa)

X_test['distance_to_versicolor'] = X_test.iloc[:,0:4].apply(distance_to_versicolor, axis=1)
X_test['distance_to_virginica'] = X_test.iloc[:,0:4].apply(distance_to_virginica, axis=1)
X_test['distance_to_setosa'] = X_test.iloc[:,0:4].apply(distance_to_setosa, axis=1)

X_test
```

Calculamos la clase de cada flor en base a la distancia euclidiana

```python
def get_class(x):
    if x['distance_to_versicolor'] < x['distance_to_virginica'] and x['distance_to_versicolor'] < x['distance_to_setosa']:
        return 'Iris-versicolor'
    elif x['distance_to_virginica'] < x['distance_to_versicolor'] and x['distance_to_virginica'] < x['distance_to_setosa']:
        return 'Iris-virginica'
    else:
        return 'Iris-setosa'
    
X_test['predicted_class'] = X_test.apply(get_class, axis=1)

X_test # Aqui podemos ver una comparativa de los resultados obtenidos con los esperados
```

Sacamos el porcentaje de aciertos

```python
X_test[X_test['class'] == X_test['predicted_class']].shape[0] / X_test.shape[0] * 100
```

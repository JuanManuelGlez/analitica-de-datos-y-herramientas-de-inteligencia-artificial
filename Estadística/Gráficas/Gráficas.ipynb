```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('multiTimeline.csv', skiprows=1)
df.head()
```

```python
# grafico de lineas
plt.figure(figsize=(10,5))
plt.plot(df['diet: (Worldwide)'], label='diet')
plt.plot(df['gym: (Worldwide)'], label='gym')
plt.plot(df['finance: (Worldwide)'], label='finance')
plt.legend()
plt.show()
```

```python
# grafico de barras del promedio de busquedas
df['diet: (Worldwide)'] = df['diet: (Worldwide)'].astype(float)
df['gym: (Worldwide)'] = df['gym: (Worldwide)'].astype(float)
df['finance: (Worldwide)'] = df['finance: (Worldwide)'].astype(float)

df['diet: (Worldwide)'] = df['diet: (Worldwide)'] / df['diet: (Worldwide)'].max()
df['gym: (Worldwide)'] = df['gym: (Worldwide)'] / df['gym: (Worldwide)'].max()
df['finance: (Worldwide)'] = df['finance: (Worldwide)'] / df['finance: (Worldwide)'].max()

plt
```

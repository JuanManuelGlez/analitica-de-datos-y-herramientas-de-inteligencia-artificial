```python
import pandas as pd
import matplotlib.pyplot as plt

# base
file_path = 'multiTimeline.csv'
df = pd.read_csv(file_path, skiprows=1)
df.columns = ['Month', 'Diet', 'Gym', 'Finance']

# Convertir Month a tipo datetime
df['Month'] = pd.to_datetime(df['Month'])

# Colores
colors = {'Diet': 'blue', 'Gym': 'red', 'Finance': 'green'}

# Gr√°fico de l√≠nea
plt.figure()
plt.plot(df['Month'], df['Diet'], label='Diet', color=colors['Diet'])
plt.plot(df['Month'], df['Gym'], label='Gym', color=colors['Gym'])
plt.plot(df['Month'], df['Finance'], label='Finance', color=colors['Finance'])
plt.title('B√∫squedas a lo largo del tiempo')
plt.xlabel('Mes')
plt.ylabel('Inter√©s de b√∫squeda')
plt.legend()
plt.show()

```

```python
# Gr√°fico de puntos
plt.figure()
plt.scatter(df['Month'], df['Diet'], label='Diet', color=colors['Diet'])
plt.scatter(df['Month'], df['Gym'], label='Gym', color=colors['Gym'])
plt.scatter(df['Month'], df['Finance'], label='Finance', color=colors['Finance'])
plt.title('Puntos de b√∫squeda')
plt.xlabel('Mes')
plt.ylabel('Inter√©s de b√∫squeda')
plt.legend()
plt.show()
```

```python
# Gr√°fico de histograma (histograma de la columna Dieta)
plt.figure()
plt.hist(df['Diet'], bins=20, color=colors['Diet'], alpha=0.7, edgecolor='black')
plt.title('Histograma de Diet')
plt.xlabel('Inter√©s de b√∫squeda')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gr√°fico de barras (promedio de b√∫squeda por categor√≠a)
mean_values = df[['Diet', 'Gym', 'Finance']].mean()
plt.figure()
mean_values.plot(kind='bar', color=[colors['Diet'], colors['Gym'], colors['Finance']], alpha=0.7, edgecolor='black')
plt.title('Promedio de inter√©s de b√∫squeda')
plt.xlabel('Categor√≠a')
plt.ylabel('Promedio de b√∫squeda')
plt.show()
```

```python
# Gr√°fico de pastel
plt.figure()
mean_values.plot(kind='pie', autopct='%1.1f%%', colors=[colors['Diet'], colors['Gym'], colors['Finance']], startangle=90, shadow=True)
plt.title('Distribuci√≥n de inter√©s de b√∫squeda')
plt.ylabel('')
plt.show()
```

## Agrega elementos adicionales para enriquecer los gr√°ficos.

Agregamos l√≠neas de cuadr√≠cula, anotaciones y el relleno debajo de la curva (fill between).

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Configuraci√≥n de tipo de letra
plt.rcParams.update({'font.size': 12, 'font.family': 'serif'})

# Colores
colors = {'Diet': 'blue', 'Gym': 'red', 'Finance': 'green'}

# Gr√°fico de l√≠nea
plt.figure()
plt.plot(df['Month'], df['Diet'], label='Diet', color=colors['Diet'], alpha=0.8)
plt.plot(df['Month'], df['Gym'], label='Gym', color=colors['Gym'], alpha=0.8)
plt.plot(df['Month'], df['Finance'], label='Finance', color=colors['Finance'], alpha=0.8)
plt.fill_between(df['Month'], df['Diet'], color=colors['Diet'], alpha=0.2)
plt.fill_between(df['Month'], df['Gym'], color=colors['Gym'], alpha=0.2)
plt.fill_between(df['Month'], df['Finance'], color=colors['Finance'], alpha=0.2)
plt.title('B√∫squedas a lo largo del tiempo')
plt.xlabel('Mes')
plt.ylabel('Inter√©s de b√∫squeda')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)
plt.show()
```

```python
# Gr√°fico de puntos
plt.figure()
plt.scatter(df['Month'], df['Diet'], label='Diet', color=colors['Diet'], alpha=0.6)
plt.scatter(df['Month'], df['Gym'], label='Gym', color=colors['Gym'], alpha=0.6)
plt.scatter(df['Month'], df['Finance'], label='Finance', color=colors['Finance'], alpha=0.6)
plt.title('Puntos de b√∫squeda')
plt.xlabel('Mes')
plt.ylabel('Inter√©s de b√∫squeda')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)
plt.show()
```

```python
# Gr√°fico de histograma
plt.figure()
plt.hist(df['Diet'], bins=20, color=colors['Diet'], alpha=0.7, edgecolor='black')
plt.title('Histograma de Diet')
plt.xlabel('Inter√©s de b√∫squeda')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# Gr√°fico de barras
mean_values = df[['Diet', 'Gym', 'Finance']].mean()
plt.figure()
mean_values.plot(kind='bar', color=[colors['Diet'], colors['Gym'], colors['Finance']], alpha=0.7, edgecolor='black')
plt.title('Promedio de inter√©s de b√∫squeda')
plt.xlabel('Categor√≠a')
plt.ylabel('Promedio de b√∫squeda')
for i, value in enumerate(mean_values):
    plt.text(i, value + 1, f'{value:.1f}', ha='center')
plt.grid(True, axis='y')
plt.show()
```

```python
# Gr√°fico de pastel
plt.figure()
mean_values.plot(kind='pie', autopct='%1.1f%%', colors=[colors['Diet'], colors['Gym'], colors['Finance']], startangle=90, shadow=True)
plt.title('Distribuci√≥n de inter√©s de b√∫squeda')
plt.ylabel('')
plt.show()
```

# PARTE 2

Usando un Dataset de Precipitaciones crear un ejemplo con 6 sub gr√°ficos (3 filas, 2 columnas) modificando al menos 5 par√°metros de visualizaci√≥n por cada subgr√°fica.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset de precipitaciones
file_path = 'pune_1965_to_2002.csv'
data = pd.read_csv(file_path)

# Asegurarse de que la columna de fechas est√© en formato datetime
data['Year'] = pd.to_datetime(data['Year'], format='%Y')

# Crear la figura y los subgr√°ficos
fig, axs = plt.subplots(3, 2, figsize=(14, 18))

# Par√°metros de visualizaci√≥n a modificar
params = {
    'title': ['January Precipitation', 'February Precipitation',
              'March Precipitation', 'April Precipitation',
              'May Precipitation', 'June Precipitation'],
    'color': ['tab:blue', 'tab:orange', 'tab:green',
              'tab:red', 'tab:purple', 'tab:brown'],
    'xlabel': ['Year'] * 6,
    'ylabel': ['Precipitation (mm)'] * 6,
    'grid': [True] * 6,
    'linestyle': ['-', '--', '-.', ':', '-', '--'],
    'marker': ['o', 's', '^', 'x', 'D', '*'],
}

# Configurar cada subgr√°fico
axs[0, 0].plot(data['Year'], data['Jan'], color=params['color'][0], linestyle=params['linestyle'][0], marker=params['marker'][0], linewidth=2)
axs[0, 0].set_title(params['title'][0])
axs[0, 0].set_xlabel(params['xlabel'][0])
axs[0, 0].set_ylabel(params['ylabel'][0])
axs[0, 0].grid(params['grid'][0])

axs[0, 1].plot(data['Year'], data['Feb'], color=params['color'][1], linestyle=params['linestyle'][1], marker=params['marker'][1], linewidth=2)
axs[0, 1].set_title(params['title'][1])
axs[0, 1].set_xlabel(params['xlabel'][1])
axs[0, 1].set_ylabel(params['ylabel'][1])
axs[0, 1].grid(params['grid'][1])

axs[1, 0].plot(data['Year'], data['Mar'], color=params['color'][2], linestyle=params['linestyle'][2], marker=params['marker'][2], linewidth=2)
axs[1, 0].set_title(params['title'][2])
axs[1, 0].set_xlabel(params['xlabel'][2])
axs[1, 0].set_ylabel(params['ylabel'][2])
axs[1, 0].grid(params['grid'][2])

axs[1, 1].plot(data['Year'], data['Apr'], color=params['color'][3], linestyle=params['linestyle'][3], marker=params['marker'][3], linewidth=2)
axs[1, 1].set_title(params['title'][3])
axs[1, 1].set_xlabel(params['xlabel'][3])
axs[1, 1].set_ylabel(params['ylabel'][3])
axs[1, 1].grid(params['grid'][3])

axs[2, 0].plot(data['Year'], data['May'], color=params['color'][4], linestyle=params['linestyle'][4], marker=params['marker'][4], linewidth=2)
axs[2, 0].set_title(params['title'][4])
axs[2, 0].set_xlabel(params['xlabel'][4])
axs[2, 0].set_ylabel(params['ylabel'][4])
axs[2, 0].grid(params['grid'][4])

axs[2, 1].plot(data['Year'], data['Jun'], color=params['color'][5], linestyle=params['linestyle'][5], marker=params['marker'][5], linewidth=2)
axs[2, 1].set_title(params['title'][5])
axs[2, 1].set_xlabel(params['xlabel'][5])
axs[2, 1].set_ylabel(params['ylabel'][5])
axs[2, 1].grid(params['grid'][5])


plt.tight_layout()
plt.show()
```

#¬†PARTE 3

Usando el Dataset de series de tiempo de tendencias de b√∫squeda de Google:
Carga un estilo (global) diferente al cl√°sico
Crea una figura donde incrustes los subgr√°ficos realizados en la primer actividad.


```python
import pandas as pd
import matplotlib.pyplot as plt

# dataset de tendencias de b√∫squeda de Google
file_path = 'multiTimeline.csv'
data = pd.read_csv(file_path, skiprows=1)
data.columns = ['Month', 'Diet', 'Gym', 'Finance']
data['Month'] = pd.to_datetime(data['Month'])


# Crear la figura y los subgr√°ficos
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Configurar cada subgr√°fico
axs[0, 0].plot(data['Month'], data['Diet'], color='tab:blue', linestyle='--', linewidth=2, marker='o')
axs[0, 0].set_title('Diet')
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Search Interest')
axs[0, 0].legend(['Diet'])

axs[0, 1].plot(data['Month'], data['Gym'], color='tab:orange', linestyle='-', linewidth=2, marker='s')
axs[0, 1].set_title('Gym')
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Search Interest')
axs[0, 1].legend(['Gym'])

axs[1, 0].plot(data['Month'], data['Finance'], color='tab:green', linestyle='-.', linewidth=2, marker='^')
axs[1, 0].set_title('Finance')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Search Interest')
axs[1, 0].legend(['Finance'])


plt.tight_layout()
plt.show()
```

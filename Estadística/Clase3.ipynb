```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_sample-2.csv')
df.head()

sales_one = df.query("QTR_ID==1")['SALES']
#print(sales_one)

sales_two = df.query("QTR_ID==2")['SALES']

sales_three = df.query("QTR_ID==3")['SALES']

sales_four = df.query("QTR_ID==4")['SALES']

sales_one
```

Podemos ver que el dataframe es una serie de registros de ventas las cuales podemos asumir que es una población, ya que no nos dan ningú motivo para pensar que es muestra, por lo que debe de ser tratada como población.

```python
#Por cada serie, estimar la meida, mediana, moda, varianza, desviación estanadar y los cuartiles

print("Q1")
print("Media ", sales_one.mean())
print("Mediana ", sales_one.median())
print("Moda ", sales_one.mode())
print("Variazanza ", sales_one.var())
print("desviación estandar ", sales_one.std(ddof=0))
print("Cuartiles \n", sales_one.quantile([0.25, 0.5, 0.75]))
print()

print("Q2")
print("Media ", sales_two.mean())
print("Mediana ", sales_two.median())
print("Moda ", sales_two.mode())
print("Variazanza ", sales_two.var())
print("desviación estandar ", sales_two.std(ddof=0))
print("Cuartiles \n", sales_two.quantile([0.25, 0.5, 0.75]))
print()

print("Q3")
print("Media ", sales_three.mean())
print("Mediana ", sales_three.median())
print("Moda ", sales_three.mode())
print("Variazanza ", sales_three.var())
print("desviación estandar ", sales_three.std(ddof=0))
print("Cuartiles \n", sales_three.quantile([0.25, 0.5, 0.75]))
print()

print("Q4")
print("Media ", sales_four.mean())
print("Mediana ", sales_four.median())
print("Moda ", sales_four.mode())
print("Variazanza ", sales_four.var())
print("desviación estandar ", sales_four.std(ddof=0))
print("Cuartiles \n", sales_four.quantile([0.25, 0.5, 0.75]))

```

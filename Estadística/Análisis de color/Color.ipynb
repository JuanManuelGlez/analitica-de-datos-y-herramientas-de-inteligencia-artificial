# Gráficas

```python
#Paletas de colores 
logo = ["#F22233", "#F2E527", "#F2D43D", "#F2A03D", "#D93814"]
monocromatico = ["#BFAE56", "#8C845B", "#F2D43D", "#59574C", "#333126"]
triada = ["#F23DC7", "#3DF2F0", "#F2D43D", "#9D925C", "#73566C"]
complementarios = ["#433DF2", "#F2D43D","#5C59B3", "#9D925C", "#575673"]
analogos = ["#F2BE3D", "#F2EA3D","#F2D43D", "#F2A83D", "#C3F23D"]
cuadrados = ["#F2493D", "#443DF2","#F2D43D", "#3DF289", "#9D925C"]
tonos = ["#CFB534", "#A69129","#F2D43D", "#7D6D1F", "#544A15"]
compuestos = ["#F2AF3D", "#A73DF2","#F2D43D", "#443DF2", "#9D925C"]
```

```python
chart1 = alt.Chart(df).mark_bar().encode(
    x=alt.X('Category', title='Categoria'),
    y=alt.Y('count()', title='Cantidad'),
    color=alt.Color("Category:N",scale=alt.Scale(range=logo), legend=None)
).properties(
    title='',
    height=400,
    width=600
)

chart1
```

```python
chart2 = alt.Chart(df).mark_arc(innerRadius=50).encode(
    theta="count:Q",
    color=alt.Color("Category:N",scale=alt.Scale(range=tonos))
).transform_aggregate(
    count='count()',
    groupby=['Category']
).transform_calculate(
    percentage="datum.count / 50"
).configure_mark(
    cornerRadius=3
).properties(
    title='Category',
    width=300,
    height=300
)

chart2
```

```python
# Pie chart of region 
alt.Chart(df).mark_arc(innerRadius=50).encode(
    theta="Region:N",
    color=alt.Color("Region:N",scale=alt.Scale(range=monocromatico)),
).properties(
    width=300,
    height=300
).transform_aggregate(
    count='count()',
    groupby=['Region']
).transform_calculate(
    percentage="datum.count / 50"
).configure_mark(
    cornerRadius=3
).configure_view(
    strokeWidth=0
).configure_title(
    fontSize=20,
    anchor='start',
    color='gray'
).configure_title(
    fontSize=20,
    anchor='start',
    color='#544A15'
).properties(
    title='Región',
    width=300,
    height=300
)
```

Podemos ver que las regiónes estan divididas en 4 perfectamente.

Al ser la suma total de las regiones podemos utilizar un grafico de pie (en este caso dona) para representarlo. Los colores elegidos fuerón los de la paleta de tonos por sus tonos sutiles.

```python
# Make radio button less cramped by adding a space after each label
# The spacing will only show up in your IDE, not on this doc page
options = ['Central', 'East', 'South', 'West']
labels = [option + ' ' for option in options]

input_dropdown = alt.binding_radio(
    # Add the empty selection which shows all when clicked
    options=options + [None],
    labels=labels + ['All'],
    name='Region: '
)
selection = alt.selection_point(
    fields=['Region'],
    bind=input_dropdown,
)

alt.Chart(df).mark_point().encode(
    x='Sales:Q',
    y='Profit:Q',
    # We need to set a constant domain to preserve the colors
    # when only one region is shown at a time
    color=alt.Color('Region:N').scale(range=logo),
).add_params(
    selection
).transform_filter(
    selection
).interactive()
```

Esta grafica **Interactiva** nos ayuda a ver la relación entre ventas y ganancia de las diferentes regiones, lo cual es muy bueno para poder hacer la comparativa y poder saber cual area es mejor y poder analizarla directamente. 

Esto nos puede ayudar en una presentación con ejecutivos por ejemplo, en donde primero mostremos las graficas de todo y despue vayamos segmentando por regiones, algo bastante util.

```python
# Crear un menú desplegable para seleccionar el eje X
dropdown = alt.binding_select(
    options=['Category', 'Region', 'Sub-Category'],
    name='X-axis column '
)
xcol_param = alt.param(
    value='Category',
    bind=dropdown
)

# Crear la gráfica de ventas
chart = alt.Chart(df).mark_circle(size=60).encode(
    x=alt.X('x:N').title(''),
    y=alt.Y('Sales:Q', title='Sales'),
    color=alt.Color('Segment:N', title='Segment', scale=alt.Scale(range=cuadrados)),
    tooltip=['Category:N', 'Region:N', 'Sub-Category:N', 'Sales:Q']
).transform_calculate(
    x=f'datum[{xcol_param.name}]'
).add_params(
    xcol_param
).properties(
    title='Sales by Selected X-axis'
).interactive()

# Mostrar la gráfica
chart.show()
```

Esta grafica esta genial, ya que son **tres** graficas en una, donde podemos ver las ventas por categoria, sub-categoria o región según nos parezca. Tambien esta coloreada por segmento, usando la paleta de **Cuadrados**. 

Podemos ver que tiene 3 dimensiones cada una de la grafica, por lo cual aporta muchisima información y nos puede ayudar a sacar buenas concluciónes al respecto.

```python
# bar chart de category y sales 
alt.Chart(df).mark_bar().encode(
    x=alt.X('Category:N', title='Category'),
    y=alt.Y('Sales:Q', title='Sales'),
    color=alt.Color('Category:N', scale=alt.Scale(range=triada)),
    tooltip=['Category:N', 'Sales:Q']
).properties(
    title='Sales by Category',
    width=500
).interactive()
```

Esta grafica nos muestra una simple comparació de las ventas por categoria utilizando los valores de Triada. Esto nos puede servir para saber cual categoría es la que tiene mas ventas. Podemos ver que esta paleta de colores nos permite el diferenciar claramente las categorías, lo cual nos puede ayudar a recalcar algún punto en el que nos ayude el tener las categorías separadas, es por eso que tambien puede ser una mejor opción que el grafico de pie.

```python
# Paletas de colores (ejemplo)
triada = ["#F23DC7", "#3DF2F0", "#F2D43D", "#9D925C", "#73566C"]
cuadrados = ["#F2493D", "#443DF2", "#F2D43D", "#3DF289", "#9D925C"]
logo = ["#F22233", "#F2E527", "#F2D43D", "#F2A03D", "#D93814"]
monocromatico = ["#BFAE56", "#8C845B", "#F2D43D", "#59574C", "#333126"]

# Gráfica de ventas con menú desplegable
dropdown = alt.binding_select(
    options=['Category', 'Region', 'Sub-Category'],
    name='X-axis column '
)
xcol_param = alt.param(
    value='Category',
    bind=dropdown
)

sales_chart = alt.Chart(df).mark_circle(size=60).encode(
    x=alt.X('x:N').title(''),
    y=alt.Y('Sales:Q', title='Sales'),
    color=alt.Color('Segment:N', title='Segment', scale=alt.Scale(range=cuadrados)),
    tooltip=['Category:N', 'Region:N', 'Sub-Category:N', 'Sales:Q']
).transform_calculate(
    x=f'datum[{xcol_param.name}]'
).add_params(
    xcol_param
).properties(
    title='Sales by Selected X-axis',
    width=500,
    height=300
).interactive()

# Gráfica de punto con filtro de región
options = ['Central', 'East', 'South', 'West']
labels = [option + ' ' for option in options]

input_dropdown = alt.binding_radio(
    options=options + [None],
    labels=labels + ['All'],
    name='Region: '
)
selection = alt.selection_point(
    fields=['Region'],
    bind=input_dropdown,
)

scatter_plot = alt.Chart(df).mark_point().encode(
    x='Sales:Q',
    y='Profit:Q',
    color=alt.Color('Region:N', scale=alt.Scale(range=logo)),
    tooltip=['Region:N', 'Sales:Q', 'Profit:Q']
).add_params(
    selection
).transform_filter(
    selection
).properties(
    title='Sales and Profit by Region',
    width=500,
    height=300
).interactive()

# Pie chart of region
pie_chart = alt.Chart(df).mark_arc(innerRadius=50).encode(
    theta=alt.Theta(field='count', type='quantitative'),
    color=alt.Color("Region:N", scale=alt.Scale(range=monocromatico)),
    tooltip=['Region:N', alt.Tooltip('count()', title='Count')]
).properties(
    title='Region',
    width=300,
    height=300
).transform_aggregate(
    count='count()',
    groupby=['Region']
)

# Bar chart of category and count
bar_chart_category = alt.Chart(df).mark_bar().encode(
    x=alt.X('Category:N', title='Category'),
    y=alt.Y('count()', title='Count'),
    color=alt.Color("Category:N", scale=alt.Scale(range=logo), legend=None),
    tooltip=['Category:N', alt.Tooltip('count()', title='Count')]
).properties(
    title='Category Count',
    height=400,
    width=600
)

# Concatenar todas las gráficas en un solo tablero
final_dashboard = alt.vconcat(
    sales_chart,
    alt.hconcat(
        scatter_plot,
        pie_chart,
        bar_chart_category
    )
).resolve_scale(
    color='independent'
).configure_mark(
    cornerRadius=3
).configure_view(
    strokeWidth=0
).configure_title(
    fontSize=20,
    anchor='start',
    color='gray'
)

# Mostrar la gráfica
final_dashboard.show()
```

El tablero interactivo combina múltiples visualizaciones para analizar las ventas de una empresa desde diferentes perspectivas. Incluye una gráfica de barras que muestra las ventas por categoría, una gráfica de burbujas con un menú desplegable que permite seleccionar diferentes dimensiones del eje X (Categoría, Región y Sub-Categoría), una gráfica de dispersión que filtra por región, un gráfico de pastel que representa la distribución de ventas por región, y una gráfica de barras que muestra el conteo de productos por categoría. Las paletas de colores variadas facilitan la diferenciación visual de las categorías y regiones.

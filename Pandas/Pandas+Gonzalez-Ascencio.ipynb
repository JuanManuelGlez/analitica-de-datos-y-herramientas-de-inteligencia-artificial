Leemos el archivo y mostramos las primeras 10 filas

```python
#Abrir el archivo y leerlo
df = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv', encoding='latin-1')
df.head(10)
```

cambiamos el formato de la fecha y ponemos hacemos que sea el indice

```python
#construir un objeto de series de tiempo con el índice igual a la fecha de los juegos olímpicos en formato tiempo
df['Date'] = pd.to_datetime(df['Year'], format='%Y')
df.set_index('Date', inplace=True) #el implace es para que se guarde en el mismo dataframe
df.head(10)
```

Separar en dos columnas el nombre del atleta

```python
#Separar nombres de apellidos en dos columnas distintas usando la coma como separador
#df['Athlete'] = df['Athlete'].str.split(',', expand=True)
#separar en dos columnas
df[['Surname', 'Name']] = df['Athlete'].str.split(' ', n=1, expand=True)
df.head(10)
```

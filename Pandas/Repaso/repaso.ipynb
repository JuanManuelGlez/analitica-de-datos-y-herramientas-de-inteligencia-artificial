(20 pts) Genera el Dataframe de la tabla Stats y realiza limpieza, y preprocesamiento de modo que:
- La columna X1 sea utilizada como índice
- Conservemos solo aquellas columnas con mas de 20,000 registros
- De este resultado, conserva sólo aquellos registros que tengan completas las columnas: Year, Age, PTS, Tm, Pos
- A partir de la variable PTS, crea otra variable Anotador que indique si el jugador genera mas de 100 puntos.
- Despliega cuántos de estos registros pertenecen a cada categoría (Anotador o no anotador).

```python
#pip install -U scikit-learn
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel('NBA.xlsx', sheet_name='stats')

#La columna X1 sea utilizada como índice
df.set_index('X1', inplace=True)
```

```python
#Conservemos solo aquellas columnas con mas de 20,000 registros
df = df.dropna(thresh=20000, axis=1)
#print(df.count())

#De este resultado, conserva sólo aquellos registros que tengan completas las columnas: Year, Age, PTS, Tm, Pos

df = df.dropna(subset=['Year', 'Age', 'PTS', 'Tm', 'Pos'])
print(df.count())
```

```python
#A partir de la variable PTS, crea otra variable Anotador que indique si el jugador genera mas de 100 puntos.

df['Anotador'] = df['PTS'] > 100

#Despliega cuántos de estos registros pertenecen a cada categoría (Anotador o no anotador).

print(df['Anotador'].value_counts())
```

(20 pts) Feature scaling: Analiza los resultados de análisis usando las técnicas de Escalamiento de esta sesión
- Genera Boxplots con sus variables numéricas (sin escalar)
- Aplica escalamiento por normalización a sus variables numéricas
- Aplica escalamiento por estandarización a sus variables numéricas
- Compara resultados en boxplots. ¿Qué podemos interpretar de cada técnica?
- Responde en un bloque de texto: ¿Qué método elegir si hay posibilidades de adquirir constantemente nuevos datos?


```python
#Genera Boxplots con sus variables numéricas (sin escalar)


pd.DataFrame(df, columns=df.columns).plot.box(figsize=(20,5), rot=90)
```

```python

#Aplica escalamiento por normalización a sus variables numéricas

X = df.select_dtypes(include='number') # seleccionar variables numericas
norm = MinMaxScaler().fit(X) # ajusta parametros de la instancia de MinMaxScaler
array_norm = norm.transform(X) # Numpy array con escalamiento segun MinMax

data_norm = pd.DataFrame(array_norm, columns = X.columns) # creo DF con datos escalados
data_norm.describe().T
```

```python
stand_scal = StandardScaler().fit(X) # ajusta parametros del imputador
array_ss = stand_scal.transform(X) # Numpy array con escalamiento segun SS

data_ss = pd.DataFrame(array_ss, columns = X.columns) # creo DF con datos escalados
data_ss.describe().T.round(3)
```

```python
#Compara resultados en boxplots. ¿Qué podemos interpretar de cada técnica?

data_norm.plot.box(figsize=(20,5), rot=90)
```

Podemos ver que la que mas información nos aporta es la de Nomralización, ya que para empezar, si no escalamos los datos no podemos tener información de valor, ya que hay muchos que estan pegados a 0 y muchos otros que se acercar a 4000

En la grafica de estdarización podemos ver que las variables estan todas mas cerca del 0 y nos da mas información de valor, pero es mas conveniente tener los datos con la grafica de normalización

Binning: Crea una copia del dataframe donde, a partir de la variable AST, generen otras variables AST_categories_(metodo) de modo que:
- Se reflejen 4 intervalos según los métodos between & loc, cut y qcut. Dichos bins deben estar basados en los cuartiles (Q1, Q2, Q3, Q4).
- Similarmente, genera mediante los mismos métodos 3 grupos: LOW, STANDARD y HIGH, que correspondan con aquellos jugadores con un desempeño en puntos (PTS) menor al percentil 10, entre 10 y 90, y mayor a 90 respectivamente.


```python
#Crea una copia del dataframe donde, a partir de la variable AST, generen otras variables AST_categories_(metodo) de modo que: Se reflejen 4 intervalos según los métodos between & loc, cut y qcut. Dichos bins deben estar basados en los cuartiles (Q1, Q2, Q3, Q4).

df['AST_categories_between'] = pd.cut(df['AST'], bins=[0, df['AST'].quantile(0.25), df['AST'].quantile(0.5), df['AST'].quantile(0.75), df['AST'].max()], labels=['Q1', 'Q2', 'Q3', 'Q4'])

df['AST_categories_loc'] = df['AST'].apply(lambda x: 'Q1' if x <= df['AST'].quantile(0.25) else 'Q2' if x <= df['AST'].quantile(0.5) else 'Q3' if x <= df['AST'].quantile(0.75) else 'Q4')

df['AST_categories_cut'] = pd.cut(df['AST'], bins=[0, df['AST'].quantile(0.25), df['AST'].quantile(0.5), df['AST'].quantile(0.75), df['AST'].max()], labels=['Q1', 'Q2', 'Q3', 'Q4'])

df['AST_categories_qcut'] = pd.qcut(df['AST'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

```

```python
#Similarmente, genera mediante los mismos métodos 3 grupos: LOW, STANDARD y HIGH, que correspondan con aquellos jugadores con un desempeño en puntos (PTS) menor al percentil 10, entre 10 y 90, y mayor a 90 respectivamente.

df['PTS_categories_between'] = pd.cut(df['PTS'], bins=[0, df['PTS'].quantile(0.1), df['PTS'].quantile(0.9), df['PTS'].max()], labels=['LOW', 'STANDARD', 'HIGH'])

df['PTS_categories_loc'] = df['PTS'].apply(lambda x: 'LOW' if x <= df['PTS'].quantile(0.1) else 'STANDARD' if x <= df['PTS'].quantile(0.9) else 'HIGH')

df['PTS_categories_cut'] = pd.cut(df['PTS'], bins=[0, df['PTS'].quantile(0.1), df['PTS'].quantile(0.9), df['PTS'].max()], labels=['LOW', 'STANDARD', 'HIGH'])

df['PTS_categories_qcut'] = pd.qcut(df['PTS'], q=3, labels=['LOW', 'STANDARD', 'HIGH'])

```

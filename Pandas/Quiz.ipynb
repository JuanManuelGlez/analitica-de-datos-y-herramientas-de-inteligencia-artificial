agregamos una nueva columna llamada GananciaUnidad que contenga el resultado de restar el precio de venta menos el precio de compra

```python
df2 = pd.read_json('DatosPractica-1.json', orient='split')
df2
```

```python
df2['GananciaUnidad'] = df2['precioVenta'] - df2['precioCompra']
df2['GananciaUnidad'].sum()
```

agregamos una nueva columna llamada IngresoTotal que calcule el ingreso  en caso de que se venda toda la existencia del producto

```python

df2['IngresoTotal'] = df2['GananciaUnidad'] * df2['existencia']
df2['IngresoTotal'].sum()
```

agregamos una nueva columna de tipo booleano, la cual se llame BajoInventario y que marque true si la existencia es menor a 40

```python
df2['BajoInventario'] = df2['existencia'] < 40
df2['BajoInventario'].sum()
df2
```

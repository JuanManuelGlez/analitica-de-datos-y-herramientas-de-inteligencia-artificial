## Loc y Iloc
* Iloc nos sirve para buscar un elemento en un Dataframe usando su indice
* Loc nos sirve para buscarlo por elemento o usar una condicional, se puede hacer una referencia a que es similar a usar un select en SQL

```python
Ajedrez_64 = np.arange(1,65).reshape(8,8)
Ajedrez_df = pd.DataFrame(
Ajedrez_64,
columns=range(1,9),
index=['A','B','C','D','E','F','G','H']
) 
Ajedrez_df
```

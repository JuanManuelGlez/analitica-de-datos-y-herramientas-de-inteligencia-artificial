 # Parte 1

 Una consultora que está desarrollando un sistema para una tienda de aparatos electrónicos, se encuentra planeando la forma de controlar los productos en un sistema de compra-venta y la estructura base que utilizará posee las siguientes clases:

En un bloque de código crea la clase Aparato que es la clase base o padre y maneja todas las cuestiones comunes a los aparatos electrónicos. Debe contener, por lo menos, los siguientes atributos:

- marca
- modelo
- precio
- ancho
- alto
- correoSoporte
- telefonoSoporte

Y los métodos que maneja son:

- calcularPrecio(porcentaje): Regresa el precio del producto con una porcentaje de ganancia aplicado al atributo precio. Por ejemplo, si el porcentaje es 10 y el precio 100, el precio que regresaría sería 110.
- obtenerDimensiones(): Regresa el ancho y alto del dispositivo en forma de tupla.
- mostrar(): Muestra en tres líneas los valores de los atributos de instancia (marca, modelo, precio, ancho, alto, correoSoporte y telefonoSoporte).



```python
class Aparato:
    def __init__(self, marca, modelo, precio, ancho, alto, correoSoporte, telefonoSoporte):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.ancho = ancho
        self.alto = alto
        self.correoSoporte = correoSoporte
        self.telefonoSoporte = telefonoSoporte

    #Metodos:
        
    def calcularPrecio(self, porcentaje):
        self.precio = self.precio + (self.precio * porcentaje / 100)
        return self.precio
    
    def obtenerDimenciones(self):
        return (self.ancho, self.alto)
    
    def mostrar (self):
        print ("Marca: ", self.marca, ", Modelo: ", self.modelo, ", Precio: ", self.precio)
        print ("Ancho: ", self.ancho, ", Alto: ", self.alto)
        print ("Correo de soporte: ", self.correoSoporte, ", Telefono de soporte: ", self.telefonoSoporte)
    
```

En otro bloque de código define el main:

Pide los siguientes valores: marca (string), modelo (string), precio (float), ancho (float) y alto (float), correo (string) y telefono (string).
Crea un objeto de la clase Aparato llamado a.
Llama al método mostrar del objeto a : a.mostrar()
Pide el porcentaje de ganancia.
Llama al método de calcularPrecio(porcentaje) e imprime el resultado.
Llama al método de obtenerDimensiones() e imprime el resultado.

```python
def main():
    marca = input("Ingrese la marca: ")
    modelo = input("Ingrese el modelo: ")
    precio = float(input("Ingrese el precio: "))
    ancho = float(input("Ingrese el ancho: "))
    alto = float(input("Ingrese el alto: "))
    correoSoporte = input("Ingrese el correo de soporte: ")
    telefono = input("Ingrese el telefono de soporte: ")

    a = Aparato(marca, modelo, precio, ancho, alto, correoSoporte, telefono)

    a.mostrar()
    porcentaje = float(input("Ingrese el porcentaje de ganancia: "))
    print ("El precio con el porcentaje de ganancia es: ", a.calcularPrecio(porcentaje))
    print ("Las dimenciones son: ", a.obtenerDimenciones())

main()
```

En un nuevo bloque de código crea la clase Tablet.  Esta clase es una clase hija de Aparato y contiene todo lo relacionado con las tablets, por ejemplo:

- procesador
- tamaño de la RAM (sizeRAM)
- tamaño del disco duro (sizeDisco)

NOTA: Definir estos tres atributos como privados.

Y los métodos que maneja son:

- Los getters y setters para manejar sus atributos privados (procesador, sizeRAM y sizeDisco)
- getUso() permite clasificar a las tablets en tres categorías:
    - Para juegos: En caso de que la memoria sea mayor o igual a 32 Gb y que el disco duro sea mayor o igual a 1000 Gigabytes.
    - Para desarrollo: En caso de que la memoria sea mayor o igual a 8 Gb o que el disco duro sea mayor o igual a 300 Gigabytes.
    - Para propósito general: En cualquier otro caso
    
Nota: El método getUso() debe regresar un string que contenga Para juegos, Para desarrollo o Para propósito general.

- mostrar(): Muestra en cuatro líneas los valores de los atributos de instancia. Utiliza los métodos getters para obtener los atributos: procesador, sizeRAM y sizeDisco. 

```python
#Clase hija de aparato llamada Tablet

class Tablet(Aparato):
    def __init__(self, marca, modelo, precio, ancho, alto, correoSoporte, telefonoSoporte, procesador, sizeRAM, sizeDisco):
        super().__init__(marca, modelo, precio, ancho, alto, correoSoporte, telefonoSoporte)
        self.procesador = procesador
        self.sizeRAM = sizeRAM
        self.sizeDisco = sizeDisco

    #Getters y Setters
        
    def getProcesador(self):
        return self.procesador
    def setProcesador(self, procesador):
        self.procesador = procesador

    def getSizeRAM(self):
        return self.sizeRAM
    def setSizeRAM(self, sizeRAM):
        self.sizeRAM = sizeRAM

    def getSizeDisco(self):
        return self.sizeDisco
    def setSizeDisco(self, sizeDisco):
        self.sizeDisco = sizeDisco

    def getUso(self):
        if self.sizeRAM >= 32 and self.sizeDisco >= 1000:
            return "Para juegos"
        elif self.sizeRAM >= 8 and self.sizeDisco >= 300:
            return "Para desarrollo"
        else:
            return "Para proposito general"
        
    def mostrar(self):
        super().mostrar()
        print ("Procesador: ", self.procesador, ", RAM: ", self.sizeRAM, ", Disco: ", self.sizeDisco)
        #print ("Uso: ", self.getUso())
```

```python
def main():
    t = Tablet("Lenovo", "LT1245", 18999, 25, 1, "psanchez@lenovo.mx", "448-335-5420","Qualcomm", 8, 256)
    t.mostrar()
    print ("El uso de la tablet es: ", t.getUso())
    print()
    
    procesador = input("Introduce el nombre del procesador: ")
    t.setProcesador(procesador)

    ram = int(input("Introduce la cantidad de RAM: "))
    t.setSizeRAM(ram)

    disco = int(input("Introduce la cantidad de disco: "))
    t.setSizeDisco(disco)

    print()

    t.mostrar()

main()
```

En un nuevo bloque de código crea la clase Verificador que permitirá validar datos antes de pasarlos a las clases y asegurar que no contendrán errores. La validación debe realizarse a través de expresiones regulares y debe tener los siguientes métodos:

Cada método recibirá un string con el valor a validar y regresará True si el valor es válido, de lo contrario regresará False. Utiliza la función match.

- verificaEmail(correo): Verificar que sea un correo electrónico válido.
    Ejemplos:

    - correo = pedro-lopez54@gmail.com regresaría True
    - correo = pedro_lopez@tec.mx regresaría True
    - correo = juansanchez.gmail regresaría False
    - correo = pedro@com regresaría False

- verificaPhone(telefono): Verificar que el teléfono se encuentre con el siguiente formato XXX-XXX-XXXX.
- verificaModel(modelo): Verificar que el modelo empiece siempre con dos letras mayúsculas y que contenga cuatro números al final.

En el mismo bloque de código crea la función valida_datos que recibe como parámetros de entrada tres strings con los valores del correo, teléfono y modelo. La función regresará True si los métodos de verificaEmail, verificaPhone y verificaModel regresan True, de lo contrario regresará False. En caso de que el resultado de los tres métodos no sea True deberá imprimir si es válido el correo, el teléfono y el modelo y deberá solicitar datos válidos con el siguiente mensaje "INTRODUCE DATOS VÁLIDOS".

```python
import re
```

```python
class Verificador: 
    def __init__(self, correo, telefono, modelo):
        self.correo = correo
        self.telefono = telefono
        self.modelo = modelo


    def verificarEmail(correo):
        if re.match('[a-zA-Z0-9.-_]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+', correo):
            return True
        else:
            return False
        
    def verificarPhone(telefono):
        if re.match('[0-9]{3}-[0-9]{3}-[0-9]{4}', telefono):
            return True
        else:
            return False
        
    def verificarModel(modelo):
        if re.match('^[A-Z]{2}[0-9]{4}$', modelo):
            return True
        else:
            return False
        
    def validaDatos(self,correo, telefono, modelo):
        if Verificador.verificarEmail(correo) and Verificador.verificarPhone(telefono) and Verificador.verificarModel(modelo):
            return True
        else:
            print("¿Es valido el correo?: ", Verificador.verificarEmail(correo))
            print("¿Es valido el telefono?: ", Verificador.verificarPhone(telefono))
            print("¿Es valido el modelo?: ", Verificador.verificarModel(modelo))
            return False
```

- Imprime un salto de línea: print()

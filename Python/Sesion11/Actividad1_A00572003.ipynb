```python
#La función limpia_archivo (nombreO, nombreD) que recibe el nombre del archivo de texto origen y el nombre del archivo de texto destino. La función debe copiar todo el contenido del archivo origen al archivo destino, cuyo contenido final será la copia de todos los caracteres válidos. Si encuentra saltos de línea copia un espacio en blanco.

def limpia_archivo(nombreO, nombreD):
    fileO = open(nombreO, "r")
    fileD = open(nombreD, "w")
    caracteres_validos =  " abcdefghijklmnopqrstuvwxyzñáéíóú0123456789ABCDEFGHIJKLMNÑOPQRSTUVWXYZÁÉÍÓÚ"
    for line in fileO:
        for char in line:
            if char in caracteres_validos:
                fileD.write(char)
            elif char == "\n":
                fileD.write(" ")
    fileO.close()
    fileD.close()

    
```

```python
#La función obtiene_palabras que regrese la lista de palabras del archivo de texto. 

def crea_lista_palabras(nombre):
    file = open(nombre, "r")
    lista = []
    for line in file:
        lista += line.split()
    file.close()
    return lista
```

```python
#La función promedio_letras que reciba la lista de palabras y regrese el promedio de letras que tienen las palabras. En el main, imprime el promedio con dos decimales.

def promedio_letras(lista):
    suma = 0
    for palabra in lista:
        suma += len(palabra)
    return round(suma/len(lista), 2)
```

```python
#La función digitos_consonantes que reciba la lista de palabras y regrese una tupla con dos valores: el primero representa la cantidad de palabras que comienzan con un digito y el segundo representa la cantidad de palabras que comienzan con una consonante (minúsculas y mayúsculas).

def digitos_consonantes(lista):
    digitos = 0
    consonantes = 0
    for palabra in lista:
        if palabra[0].isdigit():
            digitos += 1
        elif palabra[0].lower() not in "aeiou":
            consonantes += 1
    return (digitos, consonantes)
```

```python
#La función diccionario_abecedario que reciba la lista de palabras y regrese un diccionario donde las llaves son las letras del abecedario y cada elemento una lista de las palabras que comienzan con dicha letra.

def diccionario_abecedario(lista):
    letras = "abcdefghijklmnopqrstuvwxyz"
    dic = {}
    for letra in letras:
        listap = []
        for palabra in lista:
            if palabra[0].lower() == letra:
                listap.append(palabra)
        dic[letra] = listap
    return dic
```

```python
#La función imprime_diccionario que recibe el diccionario creado en la función anterior y lo imprime como una tabla, con una tabulación entre la clave y la lista. 

def imprime_diccionario(diccionario):
    for key in diccionario:
        print(key, "\t", diccionario[key])
```

```python
#La función cuenta_signos que imprime el conteo de paréntesis y signos de interrogación que se utilizaron en el archivo de texto

def cuenta_signos(nombre):
    file = open(nombre, "r")
    parentesis = 0
    signos = 0
    for line in file:
        for char in line:
            if char == "(" or char == ")":
                parentesis += 1
            elif char == "?":
                signos += 1
    print("Paréntesis:", parentesis)
    print("Signos de interrogación:", signos)
    file.close()
```

```python
def main():
    ArchivoO = "ElAlquimista.txt"
    ArchivoD = "CopiaElAlquimista.txt"
    limpia_archivo(ArchivoO, ArchivoD)

    print("LISTA DE PALABRAS")
    print(crea_lista_palabras(ArchivoD))

    print("PROMEDIO DE LETRAS")
    print(promedio_letras(crea_lista_palabras(ArchivoD)))

    print("DIGITOS Y CONSONANTES")
    print(digitos_consonantes(crea_lista_palabras(ArchivoD)))

    print("DICCIONARIO ABECEDARIO")
    print(diccionario_abecedario(crea_lista_palabras(ArchivoD)))

    print("IMPRIME DICCIONARIO")
    imprime_diccionario(diccionario_abecedario(crea_lista_palabras(ArchivoD)))

    print("CUENTA SIGNOS")
    cuenta_signos(ArchivoO)

    

main()
```
